# Объявление классов и функций

```python
def convert_date(date):
    if date == "":
        return None
    return datetime.strptime(date, '%m/%d/%Y %H:%M')

def initStation(stations):
    class Station(NamedTuple):
        station_id: int
        name: str
        lat: float
        long: float
        dockcount: int
        landmark: str
        installation: str
    
    for station in stations:
        yield Station(
            station_id = int(station[0]),
            name = station[1],
            lat = float(station[2]),
            long = float(station[3]),
            dockcount = int(station[4]),
            landmark = station[5],
            installation = datetime.strptime(station[6], '%m/%d/%Y')
        )
        
def initTrip(trips):
    class Trip(NamedTuple):
        trip_id: int
        duration: int
        start_date: datetime
        start_station_name: str
        start_station_id: int
        end_date: datetime
        end_station_name: str
        end_station_id: int
        bike_id: int
        subscription_type: str
        zip_code: str
        
    for trip in trips:
        yield Trip(                             
             trip_id = int(trip[0]),
             duration = int(trip[1] if trip[1] != "" else "0"),
             start_date = convert_date(trip[2]),
             start_station_name = trip[3],
             start_station_id = int(trip[4]),
             end_date = convert_date(trip[5]),
             end_station_name = trip[6],
             end_station_id = trip[7],
             bike_id = int(trip[8]),
             subscription_type = trip[9],
             zip_code = trip[10]
        )
```

```python
def ExtractHeadersAndData(data):
    columns = data.first().split(",")
    table = data.filter(lambda row: row != ",".join(columns)).map(lambda row: row.split(","))
    return columns, table
```

# Загрузка данных

```python
trip_data = sc.textFile("data/trips.csv")
station_data = sc.textFile("data/stations.csv")
```

```python
trip_columns, trips = ExtractHeadersAndData(trip_data)
station_columns, stations = ExtractHeadersAndData(station_data)
```

```python
stationsInternal = stations.mapPartitions(initStation)
tripsInternal = trips.mapPartitions(initTrip)
```

# 1. Найти велосипед с максимальным временем пробега

```python
# Выделим как ключ id велосипеда
trips_by_bike = tripsInternal.keyBy(lambda trip: trip.bike_id)
trips_by_bike.first()
```

```python
trips_durations = trips_by_bike\
                     .mapValues(lambda x: x.duration)\
                     .reduceByKey(lambda x1, x2: x1 + x2)
```

```python
bike_id = trips_durations.top(1, key=lambda x: x[1])[0][0]
bike_id
```

# 2. Найти наибольшее геодезическое расстояние между станциями.

```python
def DegreesToRadians(degrees):
    return degrees*math.pi/180

# https://en.wikipedia.org/wiki/Haversine_formula
def distance(lat1, long1, lat2, long2):
    earth_radius = 6371.0088
    lat1, long1 = math.radians(lat1), math.radians(long1)
    lat2, long2 = math.radians(lat2), math.radians(long2)
    
    dLat = lat2 - lat1
    dLong = long2 - long1
  
    sin_d_lat = (math.sin(dLat / 2)) ** 2
    sin_d_long = (math.sin(dLong / 2)) ** 2
    
    a = sin_d_lat + sin_d_long * math.cos(lat1) * math.cos(lat2)
  
    return 2 * earth_radius * math.atan2(math.sqrt(a), math.sqrt(1-a))
```

```python
map_distance = stationsInternal\
                               .map(lambda st: (st.station_id, st.name, st.lat, st.long))\
                               .cartesian(stationsInternal)\
                               .filter(lambda pair: pair[0][0] < pair[1][0])\ # т.к. расстояние - симметрично
                               .map(lambda pair: [pair[0][1], pair[1][1], distance(pair[0][2], pair[0][3], pair[1][2], pair[1][3])])\
                               .top(1, lambda rec: rec[2])
```

```python
distant_stations = map_distance[0]
max_distance = map_distance[0][2]

print(f"Расстояние между станциями: {distant_stations[0]} и {distant_stations[1]}, равно {max_distance:.2f} и является максимальным.")
```
